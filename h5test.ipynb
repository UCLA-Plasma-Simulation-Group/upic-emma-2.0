```python
k_array=np.arange(0,3.14, 0.001)

nk=k_array.size

print(repr(nk))

omega_analytical=k_array

omega_fdtd=np.zeros(nk)
omega_pstd=np.zeros(nk)

dt=0.71

omega_pstd=2.0/dt * np.arcsin(k_array*dt/2.0)

omega_fdtd = 2.0/dt * np.arcsin(np.sin(k_array/2.0)*dt)

SMALL_SIZE = 20
MEDIUM_SIZE = 24
BIGGER_SIZE = 28

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.figure(figsize=(12,10))
plt.plot(k_array,omega_analytical,color='black',label='Analytical/PSATD')
plt.plot(k_array,omega_fdtd,color='red',label='FDTD')
plt.plot(k_array,omega_pstd,color='blue',label='PSTD')
plt.xlabel('$k \Delta x$')
plt.ylabel('$(\omega \Delta x)/c$')
plt.legend()
plt.show()
```
